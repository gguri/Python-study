# Add new columns (that are functions of existing columns) with mutate()

### R
```R
# 새로운 변수(column) 삽입 => mutate
# 2*age한 값을 twice.age 변수로 추가
# sqrt(twice.age)한 값을 new.col 변수로 추가 
head(mutate(my.data,
           twice.age = 2*age,
           new.col = sqrt(twice.age)),
     n = 3)
```

````python
temp = data.copy()
````

````python
temp['twice_age'] = 2*temp.age
temp['new_col'] = np.sqrt(temp.twice_age)
````

````python
temp.head()
````

````python
temp1 = data.copy()
````

````python
temp1 = temp1.assign(twice_age = 2*temp1.age)
temp1 = temp1.assign(new_col = np.sqrt(temp1.twice_age))
temp1.head()
````

# Grouped operations
### R
```R
# group_by(tax) => tax 별로 grouping해서 
# sumamrise(count = n(), mean.age = mean(age,na.rm = TRUE)
# => 각 tax별 count와 age의 평균을 계산
result.1 = my.data %>%
            group_by(tax) %>%
            summarise(count = n(),
                      mean.age = mean(age, na.rm = TRUE))

head(result.1, n = 3)
```

````python
data.groupby('tax').agg({
    'tax' : {
        'tax_count' : 'count'  
    },
    'age' : {
        'mean.age' : 'mean'            
    }
}).head()
````

````python
data.groupby('tax').agg({
    'tax' : 'count',
    'age' : 'mean'
}).head()
````

````python
data.groupby('tax').agg({
    'tax' : 'count',
    'age' : 'mean'
}).head().rename(columns = {"tax" : "tax_count", "age" : "mean.age"})
````

pandas에서 groupby는 column name과 row index구조로 복잡함...ㅠ
