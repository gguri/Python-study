 # Arrange rows with arrange()

### R
```R
head(arrange(my.data, 
             age,  # 첫 번째 조건 age 
             tax), # 두 번째 조건 tax 
     n = 3)
```

````python
data.sort_values(by = ['age', 'tax']).head(3)
````

### R 
```R
head(arrange(my.data, 
             desc(age)), # age를 기준으로 내림차순 
     n = 3)
```

````python
data.sort_values(by = ['age'], ascending = False).head(3)
````

# Select columns with select()

### R
```R
head(dplyr::select(my.data,
                   nox, rm), 
     n = 3)
```

````python
data[['nox', 'rm']].head(3)
````

### R
```R
head(dplyr::select(my.data, 
                   nox:dis), 
     n = 3)
```

````python
data.loc[:, 'nox':'dis'].head(3)
````

````python
data.iloc[:, 4:8].head(3)
````
